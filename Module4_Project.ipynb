````python
# Gradient Descent
new_x = 3  # Initial value of x
gamma = 0.01  # Learning rate
precision = 0.00001  # Convergence threshold
max_iters = 10000  # Maximum number of iterations
iteration = 0  # Counter for iterations

# Define the function and its derivative
def f(x):
    return x**2 + 4*x + 6  # Example quadratic function

def df(x):
    return 2*x + 4  # Derivative of the function

while iteration < max_iters:
    previous_x = new_x
    new_x = previous_x - gamma * df(previous_x)  # [REQUIRED] Update rule for gradient descent
    step_size = abs(new_x - previous_x)

    if step_size < precision:  # [REQUIRED] Check convergence
        break

    iteration += 1

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2 + 4*x + 6

def df(x):
    return 2*x + 4

# Gradient Descent Parameters
gamma = 0.1  # Learning rate
precision = 0.0001
max_iters = 100
x_start = 3  # Initial x value

# Lists to store intermediate values for visualization
x_list = [x_start]
slope_list = [df(x_start)]

# Perform Gradient Descent
x = x_start
for _ in range(max_iters):
    grad = df(x)
    x_new = x - gamma * grad
    x_list.append(x_new)
    slope_list.append(df(x_new))
    if abs(x_new - x) < precision:  # Convergence condition
        break
    x = x_new

# Generate data for the cost function
x_1 = np.linspace(-3, 3, 500)
y_1 = f(x_1)

# Plotting
plt.figure(figsize=[20, 5])

# 1st Chart: Cost Function
plt.subplot(1, 3, 1)
plt.title("Cost Function")
plt.xlim(-3, 3)
plt.ylim(0, 20)
plt.plot(x_1, y_1, label="f(x)", color='blue')
plt.scatter(x_list, [f(i) for i in x_list], color='red', s=100, alpha=0.7, label="Steps")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

# 2nd Chart: Gradient Descent Updates
plt.subplot(1, 3, 2)
plt.title("Gradient Descent Steps")
plt.plot(x_list, [f(i) for i in x_list], color='green', label="f(x) over iterations")
plt.scatter(x_list, [f(i) for i in x_list], color='orange', s=100, alpha=0.7, label="Steps")
plt.xlabel("Iterations")
plt.ylabel("f(x)")
plt.legend()

# 3rd Chart: Slope vs x
plt.subplot(1, 3, 3)
plt.title("Slope of the Function")
plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8, label="df(x)")
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.7, label="Steps")
plt.xlabel("x")
plt.ylabel("Slope (df/dx)")
plt.legend()

plt.show()
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

### Example 4 - Data Viz with 3D Charts

#### Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

````python
## TASK-8 : Code the same things as they were done for the first two examples
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the cost function
def f(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)

# Define the partial derivatives
def fpx(x, y):
    r = 3**(-x**2 - y**2)
    return (2 * x * np.log(3) * r) / (r + 1)**2

def fpy(x, y):
    r = 3**(-x**2 - y**2)
    return (2 * y * np.log(3) * r) / (r + 1)**2

# Gradient Descent for 2D
def gradient_descent_2d(partial_fx, partial_fy, initial_guess, lr=0.1, precision=1e-6, max_iter=100):
    x, y = initial_guess
    path_x, path_y, path_cost = [x], [y], [f(x, y)]

    for _ in range(max_iter):
        grad_x = partial_fx(x, y)
        grad_y = partial_fy(x, y)
        x -= lr * grad_x
        y -= lr * grad_y

        path_x.append(x)
        path_y.append(y)
        path_cost.append(f(x, y))

        # Stop if change in cost is minimal
        if len(path_cost) > 1 and abs(path_cost[-1] - path_cost[-2]) < precision:
            break

    return (x, y), path_x, path_y, path_cost

# Initial guess
initial_guess = (2, 2)

# Run gradient descent
(local_min, path_x, path_y, path_cost) = gradient_descent_2d(fpx, fpy, initial_guess, lr=0.1)

# Visualization
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

fig = plt.figure(figsize=(16, 8))

# 3D Plot of the cost function
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.scatter(path_x, path_y, path_cost, color='red', label='Gradient Descent Path')
ax1.set_title('Cost Function Surface')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')
ax1.legend()

# 2D Projection with gradient descent steps
ax2 = fig.add_subplot(122)
ax2.contourf(X, Y, Z, levels=50, cmap='viridis', alpha=0.8)
ax2.plot(path_x, path_y, color='red', marker='o', label='Gradient Descent Path')
ax2.set_title('2D Contour and Gradient Descent Path')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
